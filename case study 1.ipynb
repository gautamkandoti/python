```python
#16.Find:
#a.	Mean and std deviation of the income
#b.	For income greater than or equal to 10.5, find the mean income
#c.	For each brand having price >=2.5 find the mean, median, sd of the log of income

print(study.INCOME.mean())
print(stat.pstdev(study.INCOME))

print(study[study.INCOME>=10.5].INCOME.mean())

print(study[study.price>=2.5].groupby(['brand']).INCOME.mean())
print(study[study.price>=2.5].groupby(['brand']).INCOME.median())
print(study[study.price>=2.5].groupby(['brand']).INCOME.std())
```
